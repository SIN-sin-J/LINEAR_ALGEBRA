```python
import sympy as sp
import matplotlib.pyplot as plt

A = sp.Matrix([[1, 6], [5, 2]])
U = sp.Matrix([3, -2])
V = sp.Matrix([6, -5])
AU=A*U
print('AU=',AU)
AV=A*V
print('AV=',AV)
#checks the vectirs" each row if eigen or not
scalar_U = all(AU[i] / U[i] == AU[0] / U[0] for i in range(len(U)))
scalar_V = all(AV[i] / V[i] == AV[0] / V[0] for i in range(len(V)))
print(f'Is V={V} an eigen vector ? {scalar_V}')
print(f'Is U={U} an eigen vector ? {scalar_U}')

plt.figure().set_size_inches(5, 5)

# U and V PLOTTING
plt.plot([0, U[0]], [0, U[1]], 'y', label='U')
plt.plot([0, V[0]], [0, V[1]], 'g', label='V')

# AU and AV PLOTTING
plt.plot([0, AU[0]], [0, AU[1]], 'r', label='AU')
plt.plot([0, AV[0]], [0, AV[1]], 'b', label='AV')

# LABELING
for vector, label in zip([U, V, AU, AV], ['U', 'V', 'AU', 'AV']):
    plt.text(vector[0], vector[1], label)

# AXIS AND LABELS
plt.xlabel('X axis')
plt.ylabel('Y axiz')
plt.title('EIGEN VECTOR COMPARISON')
plt.legend()
plt.show()
```
